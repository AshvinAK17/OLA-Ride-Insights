```python
import pandas as pd
path = r"C:\Ashvin\AI ML\Data\OLA_DataSet.xlsx"
df = pd.read_excel(path)
df
```

```python
# TO identify the null values
df.isna().sum()
```

```python
# Replacing the null values with its appropriate values based on dataset ( because the null values are only for the 
# rides that are cancelled )

df = df.fillna(value={
    'Canceled_Rides_by_Customer': 'Not Applicable',
    'Canceled_Rides_by_Driver': 'Not Applicable',
    'Incomplete_Rides_Reason' : 'Not Applicable',
    'V_TAT': 0,
    'C_TAT': 0,
    'Driver_Ratings' : 0,
    'Customer_Rating': 0,
    'Payment_Method': 'None',
    'Incomplete_Rides': 'Yes',
})
df
```

```python
#Ensuring no null values are present
df.isna().sum()
```

```python
#Standardized text data by converting to uppercase, formatted the date column, and removed irrelevant columns for 
# analysis.
df['Payment_Method'] = df['Payment_Method'].str.strip().str.upper()
df['Date'] = df['Date'].apply(lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else None )
df.drop('Vehicle Images',axis=1,inplace=True)
df['Booking_Status'] = df['Booking_Status'].str.strip().str.upper()
df['Time'] = df['Time'].apply(lambda x: x.strftime('%H:%M:%S') if pd.notnull(x) else None )
```

```python
# 1. Retrieve all successful bookings:
import matplotlib.pyplot as plt

query = """ 
SELECT * 
FROM ola_rides 
WHERE Booking_Status = 'SUCCESS'
"""
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
succesful_bookings = pd.DataFrame(results,columns=columns)

# Convert 'Date' column to datetime if needed
succesful_bookings['Date'] = pd.to_datetime(succesful_bookings['Date'])

# Group by Date
daily_counts = succesful_bookings.groupby('Date').size()

# Plot
plt.figure(figsize=(12, 6))
daily_counts.plot(kind='line', marker='o', color='green')
plt.title("Successful Bookings Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Bookings")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. The average ride distance for each vehicle type:
query = """
SELECT Vehicle_Type, AVG(Ride_Distance) AS Avg_Ride_Distance
FROM ola_rides
WHERE Booking_Status = 'SUCCESS'
GROUP BY Vehicle_Type;
"""
cursor.execute(query)
results = cursor.fetchall()
for row in results:
    print(f"Vehicle Type: {row[0]}, Average Ride Distance: {row[1]:.2f} km")
```

```python
#vefiying the data for 2nd query
success_df = df[df['Booking_Status'] == 'SUCCESS']
Avg_Distance = success_df.groupby('Vehicle_Type')['Ride_Distance'].mean()
Avg_Distance
```

```python
#3. the total number of cancelled rides by customers:
query = """
SELECT COUNT(*) AS Total_Customer_Cancelled_Rides
FROM ola_rides
WHERE Canceled_Rides_by_Customer != 'Not Applicable';
"""
cursor.execute(query)
results = cursor.fetchone()
print("Total Cancelled Rides by Customers:", results[0])
```

```python
#verifying the data for 3rd query
cancel_by_customer = df[df['Canceled_Rides_by_Customer'] != 'Not Applicable'].shape[0]
print("Total Cancelled by Customer:", cancel_by_customer)
```

```python
# 4. List the top 5 customers who booked the highest number of rides:
query = """
SELECT 
    TRIM(Customer_ID) AS CUSTOMERS,
    COUNT(*) AS TOTAL_RIDES
FROM ola_rides
WHERE Booking_Status = 'SUCCESS'
GROUP BY TRIM(Customer_ID)
ORDER BY TOTAL_RIDES DESC, TRIM(Customer_ID) ASC
LIMIT 5;
"""
cursor.execute(query)
results = cursor.fetchall()

for customer in results:
    print("Customer:", customer[0],"| Total Rides:",customer[1])
```

```python
#verifying the data for 4th query
top_customers = df[df['Booking_Status'] == 'SUCCESS']\
                    .groupby('Customer_ID')\
                    .size()\
                    .reset_index(name='Total_Rides')\
                    .sort_values(by=['Total_Rides', 'Customer_ID'], ascending=[False, True])\
                    .head(5)
top_customers
```

```python
# 5. Get the number of rides cancelled by drivers due to personal and car-related issues:
query = """
SELECT COUNT(*) AS Canceled_Rides_by_Driver
FROM ola_rides
WHERE Canceled_Rides_by_Driver = 'Personal & Car related issue';
"""
cursor.execute(query)
results = cursor.fetchone()
print("Total Cancelled Rides by drivers due to personal and car-related issues:", results[0])
```

```python
#verifying the data for 5 th query
cancelled_rides = df[df['Canceled_Rides_by_Driver'] == 'Personal & Car related issue'].shape[0]
cancelled_rides 
```

```python
#6. Find the maximum and minimum driver ratings for Prime Sedan bookings:
query = """
SELECT 
    MAX(Driver_Ratings),
    MIN(Driver_Ratings)
FROM ola_rides
WHERE Vehicle_Type = 'Prime Sedan' AND Driver_Ratings >= 1
"""
cursor.execute(query)
results = cursor.fetchone()

print("The Maximum driver ratings for Prime Sedan Booking:", results[0]) 
print("The Minimum driver ratings for Prime Sedan Booking:", results[1])
```

```python
#verifying the data for 6th query 
max_rating = df['Driver_Ratings'].max()
min_rating = df[df['Driver_Ratings']>0]['Driver_Ratings'].min()
print("The Maximum driver ratings for Prime Sedan Booking:", max_rating) 
print("The Minimum driver ratings for Prime Sedan Booking:", min_rating)
```

```python
# 7. Retrieve all rides where payment was made using UPI:

query = """
SELECT *
FROM ola_rides
WHERE Payment_Method = 'UPI'
"""
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
upi_df = pd.DataFrame(results, columns=columns)

#converting 'Date' column to datetime 
upi_df['Date'] = pd.to_datetime(upi_df['Date'])
upi_trend = upi_df.groupby('Date').size()

plt.figure(figsize=(12, 6))
upi_trend.plot(kind='line', marker='o', color='purple')
plt.title("UPI Rides Over Time")
plt.xlabel("Date")
plt.ylabel("Number of UPI Bookings")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 8. Find the average customer rating per vehicle type:

query = """
SELECT Vehicle_Type, AVG(Customer_Rating)
FROM ola_rides
GROUP by Vehicle_Type
ORDER BY AVG(Customer_Rating)
"""

cursor.execute(query)
results= cursor.fetchall()
for row in results:
    print(f"The average customer rating for {row[0]} is {row[1]:2f}")
```

```python
#verifying the 8th query
print(df.groupby('Vehicle_Type')['Customer_Rating'].mean())
```

```python
# 9. Calculate the total booking value of rides completed successfully:
query ="""
SELECT SUM(Booking_Value)
FROM ola_rides
WHERE Booking_Status = 'SUCCESS' AND Incomplete_Rides = 'No'
"""
cursor.execute(query)
results = cursor.fetchone()
print(f"THe Total value of Successfull Bookings is ₹{results[0]:,.2f}")
```

```python
#verifying the data for 9th query
filtered_df = df[(df['Booking_Status'] == 'SUCCESS') & (df['Incomplete_Rides'] == 'No')]
output = filtered_df['Booking_Value'].sum()
output 
```

```python
# 10. List all incomplete rides along with the reason
query = """
SELECT Incomplete_Rides, Incomplete_Rides_Reason
FROM ola_rides
WHERE Incomplete_Rides = 'Yes'
"""
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
incomplete_ride = pd.DataFrame(results,columns=columns)

#Filter only incomplete rides with valid reasons
filtered_df = df[
    (df['Incomplete_Rides'] == 'Yes') & 
    (df['Incomplete_Rides_Reason'] != 'Not Applicable')
]

#Group by Vehicle_Type and Reason
grouped = filtered_df.groupby(['Vehicle_Type', 'Incomplete_Rides_Reason']).size().unstack(fill_value=0)

grouped.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab10')
plt.title("Incomplete Rides by Reason & Vehicle Type (Excl. 'Not Applicable')")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
